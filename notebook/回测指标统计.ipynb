```python
files = [
    # r"D:\code\forecast_model\notebook\20241018_1_filter_0\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241018_1_filter_1\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241019_1_filter_2\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241021_1_depth_top10_20\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241021_1_god_label_depth\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241020_1_depth_2class\backtest_results.csv",
    # r'D:\code\forecast_model\notebook\20241012_1_base\backtest_results.csv',
    # r'D:\code\forecast_model\notebook\20241020_1_base_time\backtest_results.csv',
    # r'D:\code\forecast_model\notebook\20241022_1_depth_clear_keep_dup\backtest_results.csv',
    # r'D:\code\forecast_model\notebook\20241022_1_depth_passn\backtest_results.csv',
    # r'D:\code\forecast_model\notebook\20241022_1_depth_add_deal\backtest_results.csv',
    # r"D:\code\forecast_model\notebook\20241022_1_depth_deal_model\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241022_1_depth_order_nodiv\backtest_results.csv"
    # r"D:\code\forecast_model\notebook\20241026_1_model_stack\backtest_results.csv"
    # r"D:\code\forecast_model\notebook\20241024_1_depth_part_data\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241027_1_base_deeplob\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241027_1_base_part_bin\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241028_1_model_stack\backtest_results.csv"
    # r"D:\code\forecast_model\notebook\20241028_1_base_deeplob_mid_std\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241029_1_model_stack\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241029_1_model_stack_2\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241030_1_base_axiallob\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241031_1_model_stack\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241031_1_binctabl_max_nor\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241101_1_deeplob_dropout\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241101_1_of_bincatbl\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241101_1_of_deeplob\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241101_1_depth_deeplob\backtest_results.csv",
    # R"D:\code\forecast_model\notebook\20241102_1_depth_symbol\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241102_1_depth_symbol\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241103_1_base_symbol\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241103_1_depth_deal_symbol\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241103_1_depth_order_symbol\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241103_1_of_symbol\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241105_1_of_bincatbl\backtest_results.csv",
    # r"D:\code\forecast_model\notebook\20241105_1_of_symbol\backtest_results.csv"
    # r"D:\code\forecast_model\notebook\20241105_1_of_symbol_0.5\backtest_results.csv"
    # r"D:\code\forecast_model\notebook\20241105_1_of_label_1_cls\backtest_results.csv"
    r"D:\code\forecast_model\notebook\20241031_1_model_stack\backtest_results.csv"
]

data = pd.DataFrame()
for file in files:
    _data = pd.read_csv(file, dtype={'code': str})

    # 只使用 top 数据集
    # _data = _data.loc[_data['title'].str.contains('top5')].reset_index(drop=True)
    _data.sort_values(by=('information_ratio'), ascending=False)

    # max 标记
    _data['title'] = _data['title'] + _data['tag'].apply(lambda x: '_max' if 'max' in str(x) else '')

    # 筛选code
    _data = _data.loc[_data['code'].isin(['518880', '513050']), :].reset_index(drop=True)

    # # 只使用 final/max 数据
    # _data = _data.loc[~_data['title'].str.contains('max')].reset_index(drop=True)
    # _data = _data.loc[_data['title'].str.contains('final')].reset_index(drop=True)
    
    # 多seed平均
    _data = _data.loc[:, ['title', 'code', 'information_ratio', 'sharpe_ratio']]
    title = _data['title']
    _data['title'] = _data['title'].apply(lambda x: '_'.join([i for i in x.split('_') if 'seed' not in i]))
    _data['title2'] = title

    data = pd.concat([data, _data], ignore_index=True)

# 删除其中的 code (6位数字)
# 多code平均
def remove_code(input_str):
    matches = re.findall(r'_\d{6}', input_str)
    for match in matches:
        input_str = input_str.replace(match, '')
    return input_str
data['title'] = data['title'].apply(remove_code)

data.sort_values(by=(['information_ratio', 'sharpe_ratio']), ascending=False).iloc[:20]
```
