```python
from snake_env import SnakeEnv
import matplotlib.pyplot as plt
```

```python
env = SnakeEnv({})
obs, info = env.reset()
print(obs.shape)

plt.imshow(obs)
plt.show()
```

```python
plt.figure(figsize=(15,5))
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(obs[:,:,i])
    plt.axis('off')
    plt.title(f'Channel {i}')
plt.show()
```

```python
# 0 右
# 1 左
# 2 下
# 3 上
obs, reward, terminated, truncated, info = env.step(3)
print(reward, info['score'])

plt.imshow(obs)
plt.show()
```
